#### Degree Centrality

Number of links held by each node. Simplest measure of node connectivity. 
    
Used for finding very connected individuals, popular individuals, individuals who are likely to hold most information or individuals who can quickly connect with the wider network.


```python
most_popular = nx.degree_centrality(G_pol)

for w in sorted(most_popular, key=most_popular.get, reverse=True):
    print(w, most_popular[w])
```

##### Shri Conrad Kongkal Sangma, the Chief minister of Meghalaya followed by Shri Jairam Thakur,  the Chief minister of Himachal Pradesh are the most popular amoung the Chief ministers  

#### EigenVector Centrality <br>

It decides that a node is important if it is connected to other important nodes.

```python
most_important = nx.eigenvector_centrality(G_pol)

for w in sorted(most_important, key=most_important.get, reverse=True):
    print(w, most_important[w])
```

##### Shri Shivraj Singh Chouhan, the Chief minister of Madhya Pradesh followed by Shri Sarbananda Sonowal, the Chief minister of Assam are the most important amoung the  Chief ministers  

#### PageRank

PageRank is a variant of EigenCentrality, also assigning nodes a score based on their connections, and their connectionsâ€™ connections. The difference is that PageRank also takes link direction and weight into account â€“ so links can only pass influence in one direction, and pass different amounts of influence.

This measure uncovers nodes whose influence extends beyond their direct connections into the wider network.

Fun Fact: PageRank is famously one of the ranking algorithms behind the original Google search engine.

```python
pagerank = nx.pagerank(G_pol, alpha=0.9)
for w in sorted(pagerank, key=pagerank.get, reverse=True):
    print(w, pagerank[w])
```

##### Shri Shivraj Singh Chouhan, the Chief minister of Madhya Pradesh followed by Shri Manohar Lal, the Chief minister of Haryana are the most important amoung the  Chief ministers

#### Betweenness centrality 

It measures the number of times a node lies on the shortest path between other nodes.

This measure shows which nodes are â€˜bridgesâ€™ between nodes in a network. 
It does this by identifying all the shortest paths and then counting how many times each node falls on one.



```python
best_connector = nx.betweenness_centrality(G_pol)
for w in sorted(best_connector, key=best_connector.get, reverse=True):
    print(w, best_connector[w])
```

##### Shri Conrad Kongkal Sangma, the Chief minister of Meghalaya followed by Shri Shivraj Singh Chouhan, the Chief minister of Madhya Pradesh are acting as a bridge between the  Chief ministers 

```python
pos = nx.spring_layout(G_pol)
betCent = nx.betweenness_centrality(G_pol, normalized=True, endpoints=True)
node_color = [20000.0 * G_pol.degree(v) for v in G_pol]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(8,8))
nx.draw_networkx(G_pol, pos=pos, with_labels=True,
                 node_color=node_color,
                 node_size=node_size)
plt.axis('off')
```

#### Closeness centrality 

It scores each node based on their â€˜closenessâ€™ to all other nodes in the network.

Used for finding the individuals who are best placed to influence the entire network most quickly.

```python
most_influential  = nx.closeness_centrality(G_pol)
for w in sorted(most_influential , key=most_influential.get, reverse=True):
    print(w, most_influential [w])
```

##### Shri Shivraj Singh Chouhan, the Chief minister of Madhya Pradesh followed by  Shri Sarbananda Sonowal, the Chief minister of Assam are the most influential amoung the  Chief ministers  

#### Building A Subgroup 

Retrieve all the nodes connected to a given node within a NetworkX graph 

Method used : Breadth-first search

```python
group1 = nx.bfs_tree(G_pol,"AmitShah")
group2 = nx.bfs_tree(G_pol,"rajnathsingh")
group3 = nx.bfs_tree(G_pol,"narendramodi")
```

```python
plt.figure(figsize=(15,15))
nx.draw_networkx(group1, with_labels=True, node_color='pink', node_size=700, alpha=0.5, arrows=True)
plt.show()
```

```python
plt.figure(figsize=(15,15))
nx.draw_networkx(group2, with_labels=True, node_color='yellow', node_size=700, alpha=0.5, arrows=True)
plt.show()
```

```python
plt.figure(figsize=(15,15))
nx.draw_networkx(group3, with_labels=True, node_color='r', node_size=700, alpha=0.5, arrows=True)
plt.show()
```

#### Clustering 

It is observed that people who share connections in a social network tend to form associations. 

In other words, there is a tendency in a social network to form clusters.

##### Two types of Clustering <br>
Local Clustering Coefficient <br>
Global Clustering Coefficient <br>

##### Local Clustering Coefficient

Fraction of pairs of the node's friends that are friends with each other. 

##### Global Clustering

Measures the Clustering of the whole network by measuring the overall Average local clustering coefficient of all nodes in a graph


```python
# Local Clustering Coefficient 
print("Local Clustering Coefficient for narendramodi: ",nx.clustering(G_pol,'narendramodi'))
print("Local Clustering Coefficient for rajnathsingh: ",nx.clustering(G_pol,'rajnathsingh'))

# Global Clustering
print("Global clustering using Average clustering: ", nx.average_clustering(G_pol))
```
